# Taking Photos Using OpenCV

This notebook uses OpenCV and a webcam to take a photo every second, intended to be used to regularly taking photos of Ayalon Road, processing the information of the vehicles location, and save it compactly.

This direction was abandoned in favor of taking HyperLapse videos with a smartphone.

```python
import numpy as np
import scipy as sp
import scipy.stats as stats
import matplotlib as mpl
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit
import pandas as pd
import seaborn as sns
import re
from pathlib import Path
import pickle as pkl
from time import time
from tqdm import tqdm, tnrange, tqdm_notebook
from pprint import pprint
import os, sys
from warnings import warn
import cv2
```

```python
DATA = Path('data')
```

```python
ret, frame = cv2.VideoCapture(0).read()
plt.imshow(frame)
```

```python
cam = cv2.VideoCapture(0)

cv2.namedWindow("test")

img_counter = 0

while True:
    ret, frame = cam.read()
    cv2.imshow("test", frame)
    if not ret:
        break
    k = cv2.waitKey(1) # use sleep 

    if k%256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
    elif k%256 == 32:
        # SPACE pressed
        img_name = str(DATA/f"opencv_frame_{img_counter:d}.png")
        cv2.imwrite(img_name, frame)
        print("{} written!".format(img_name))
        img_counter += 1

cam.release()

cv2.destroyAllWindows()
```
